```python
# Problema 1:
""" Un hospital ha desarrollado un sistema de detección rápida para identificar si un paciente presenta 
sepsis. El sistema es bastante confiable: el 95% de los pacientes con sepsis son correctamente 
identificados como positivos, y el 95% de los pacientes sanos son correctamente identificados 
como negativos. Se sabe por estadísticas clínicas que exactamente 1 de cada 100 pacientes que 
ingresan a urgencias tiene sepsis. Un paciente recién ingresado es evaluado por el sistema y el 
resultado es "POSITIVO PARA SEPSIS". ¿Cuál es la probabilidad de que este paciente 
realmente tenga sepsis? Muestre su desarrollo. """

def calcular_probabilidad_sepsis() -> float:
    # Calcula P(S|+) usando el Teorema de Bayes.
    prevalencia = 0.01     # Probabilidad de tener sepsis (P(S))
    sensibilidad = 0.95    # Probabilidad de test POSITIVO dado que el paciente tiene sepsis (P(+|S))
    especificidad = 0.95   # Probabilidad de test NEGATIVO dado que el paciente no tiene sepsis (P(-|¬S))

    falso_positivo = 1 - especificidad # Probabilidad de test POSITIVO dado que el paciente no tiene sepsis (P(+|¬S))
    no_sepsis = 1 - prevalencia # Probabilidad de estar sano (P(¬S))

    # Aplicando el Teorema de Bayes:
    # numerador = P(+|S) * P(S)
    numerador = sensibilidad * prevalencia   # Verdaderos positivos P(+|S) * P(S)
    # denominador = numerador + P(+|¬S) * P(¬S)
    denominador = numerador + (falso_positivo * no_sepsis) # P(+|S) * P(S) + P(+|¬S) * P(¬S)

    return numerador / denominador # probabilidad de tener sepsis dado positivo P(S|+)

####################################################3
# Problema 2:
"""" Una señal de ECG en reposo presenta un voltaje que se distribuye de manera uniforme en el 
intervalo [a, b] mV, con función de densidad de probabilidad: f(x) = 1/(b-a) para a ≤ x ≤ b, y f(x) = 0 en otro caso. 
Calcule la media y la varianza de esta señal. Muestre su desarrollo."""

# Media de una distribución uniforme: E[X] = ∫[a,b] x * (1 / (b - a)) dx = (a + b) / 2
# Varianza de una distribución uniforme: Var[X] = ∫[a,b] (x - E[X])² * (1 / (b - a)) dx = ((b - a) ** 2) / 12

def uniforme_stats(a: float, b: float):
    # Calcula media y varianza de U(a, b).
    if a >= b:
        raise ValueError("Se requiere a < b")

    media = (a + b) / 2
    varianza = ((b - a) ** 2) / 12

    return media, varianza

################# MAIN #################
def main() -> None:
    #Problema 1
    resultado = calcular_probabilidad_sepsis()
    print("Problema 1:");
    print(f"   - Probabilidad de tener sepsis dado positivo: {resultado:.4f} ({resultado * 100:.2f}%)\n")
    
    # Problema 2
    print("Problema 2:");
    a, b = 0.0, 1.0 # Ejemplo de intervalo [a, b] de la señal ECG
    media, varianza = uniforme_stats(a, b)
    print(f"   - Media: {media:.4f}")
    print(f"   - Varianza: {varianza:.4f}")

if __name__ == "__main__":
    main()
```
